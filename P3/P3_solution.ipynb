```python
import numpy as np
from collections import defaultdict

def load_graph(file_path):
    edges = set()
    nodes = set()
    
    with open(file_path, 'r') as f:
        for line in f:
            src, dst = map(int, line.strip().split())
            edges.add((src, dst))  # remove duplicates automatically
            nodes.add(src)
            nodes.add(dst)
    
    return edges, nodes


def build_adj_list(edges):
    adj = defaultdict(list)
    
    for src, dst in edges:
        adj[src].append(dst)
    
    return adj


def initialize_ranks(nodes):
    n = len(nodes)
    ranks = {node: 1.0 / n for node in nodes}
    return ranks, n


def compute_pagerank(adj, ranks, nodes, n, beta=0.8, iterations=40):
    
    for _ in range(iterations):
        new_ranks = {node: (1 - beta) / n for node in nodes}
        
        for node in nodes:
            if node in adj:
                out_links = adj[node]
                share = ranks[node] / len(out_links)
                
                for dest in out_links:
                    new_ranks[dest] += beta * share
        
        ranks = new_ranks
    
    return ranks


def get_top_bottom(ranks):
    sorted_nodes = sorted(ranks.items(), key=lambda x: x[1], reverse=True)
    
    top5 = sorted_nodes[:5]
    bottom5 = sorted_nodes[-5:]
    
    return top5, bottom5
```

```python
small_path = "../datasets/Q3- PageRank/small.txt"

edges, nodes = load_graph(small_path)
adj = build_adj_list(edges)
ranks, n = initialize_ranks(nodes)

ranks = compute_pagerank(adj, ranks, nodes, n)

top5, bottom5 = get_top_bottom(ranks)

print("===== SMALL GRAPH =====")
print("Top 5:", top5)
print("Bottom 5:", bottom5)
print("\nTop score:", top5[0][1])
```

```python
full_path = "../datasets/Q3- PageRank/whole.txt"

edges, nodes = load_graph(full_path)
adj = build_adj_list(edges)
ranks, n = initialize_ranks(nodes)

ranks = compute_pagerank(adj, ranks, nodes, n)

top5, bottom5 = get_top_bottom(ranks)

print("\n===== FULL GRAPH RESULTS =====")

print("\nTop 5 nodes with highest PageRank:")
for node, score in top5:
    print(f"Node {node}: {score:.6f}")

print("\nBottom 5 nodes with lowest PageRank:")
for node, score in bottom5:
    print(f"Node {node}: {score:.6f}")
```
